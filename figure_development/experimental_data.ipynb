```python
import sys
import os
import numpy as np
import pickle
import pandas as pd
import seaborn as sns
import sklearn
import matplotlib.pyplot as plt
import itertools
import qinfer as qi
from lfig import LatexFigure

sys.path.append("/home/bf16951/QMD")
import qmla

%matplotlib inline
```

```python
experimental_data_path = os.path.join(
    "..", "experimental_study", "figures", "data", "NVB_rescale_dataset.p"
)
```

```python
exp_data = pickle.load(open(experimental_data_path, 'rb'))
```

```python
times = sorted(exp_data.keys())
likelihoods = [
    exp_data[t] for t in times
]
```

```python
lf = LatexFigure(fraction=0.5, auto_label=False)
ax = lf.new_axis()

ax.scatter(
    times, 
    likelihoods,
    s = 3,
    c= 'red',
)
ax.set_ylabel(
    r"$\Pr(0 | \hat{H}_0, t)$", 
)
ax.set_xlabel(r"$t (\mu s)$ ")
save_path = os.path.join(
    "..", "experimental_study", "figures", "raw_data.pdf", 
)

lf.save(save_path)
```
