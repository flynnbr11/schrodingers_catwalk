```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from latex_figure import LatexFigure
```

```python
lf = LatexFigure(
    fraction = 0.75, 
    use_gridspec=True, 
    gridspec_layout=(3, 2),
    gridspec_params = {
#         'hspace' : 1,
#         'wspace' : 2
    },
    font_scale=1.,
    plot_style='seaborn'
)

ax1 = lf.new_axis()
ax2 = lf.new_axis()
ax3 = lf.new_axis(ax_params={'sharex' : ax1})
ax4 = lf.new_axis()
ax5 = lf.new_axis(ax_params={'sharex' : ax3})
ax6 = lf.new_axis(ax_params={'sharex' : ax4})


x1 = np.arange(0, 10, 0.1)
x2 = np.arange(0, 100, 1)
ax1.plot(x1, np.sin(x1))
ax2.plot(x2, np.sin(x2))

x3 = np.arange(0, 5, 0.1)
x4 = np.arange(0, 7, 1)
ax5.plot(x3, np.sin(x3), )
ax6.plot(x4, np.sin(x4))

ax3.plot(x3, np.cos(x3), )
ax4.plot(x4, np.cos(x4))

# ax1.set_xticks([])
# ax2.set_xticks([])
# ax3.set_xticks([])
# ax4.set_xticks([])
ax5.set_xticks([0,5,10])
plt.setp(ax1.get_xticklabels(), visible=False)
plt.setp(ax3.get_xticklabels(), visible=False)


lf.save('imp_test_gs_3x2.png')
```
